# tools

```python
from langchain.tools import tool
```

```python
import requests
```

```python
@tool
def get_weather(city: str) -> str:
    """Get weather for a city and according to the degree create ur own condition"""

    data = requests.get(f"https://wttr.in/{city}?format=j1").json()

    temp = data["current_condition"][0]["temp_C"]

    return f"{city}: {temp}°C"
```

```python
weather_agent = create_agent(
    model="groq:llama-3.3-70b-versatile",
    system_prompt="you are a weather agent , who tell weather for the reagions",
    tools=[get_weather],
)
```

```python
weather_agent.invoke({"messages":"hi who are you"})
```

```python
resposne=weather_agent.invoke({"messages": "hi what is the weather in chennai"})
```

```python
resposne["messages"][-1].content
```

```python
resposne
```

# strctured output

```python
from pydantic import BaseModel,Field
```

```python
class weather_response(BaseModel):
    city: str = Field(description="user asked city")
    temp : str = Field(description="temperature of the city")
    condition: str = Field(description="what is the condition of the city like hot ,cool, warm")
    suggestion:str = Field(description="what you will suggest for this temperature")
```

```python
strctured_weather_agent = create_agent(
    model="groq:llama-3.3-70b-versatile",
    system_prompt="you are a weather agent , who tell weather for the reagions",
    tools=[get_weather],

)
```

```python
response=strctured_weather_agent.invoke({"messages": "hi what is the weather in kashmir"})
```

```python
response["structured_response"]
```

```python
response = strctured_weather_agent.invoke(
    {"messages": "what is my name"}
)
```

```python
response["messages"][-1].content
```

# Memory (short term)

```python
from langgraph.checkpoint import sqlite

from sqlite3 import connect
```

```python
conn=connect("memory.db",check_same_thread=False)
```

```python
checpointer=sqlite.SqliteSaver(conn)
```

```python
mem_agent = create_agent(
    model="groq:llama-3.3-70b-versatile",
    system_prompt="you are a weather agent , who tell weather for the reagions",
    tools=[get_weather],
    checkpointer=checpointer
)
```

```python
config = {"configurable": {"thread_id": "u2"}}
```

```python
response=mem_agent.invoke({"messages":"hi i am hari"},config=config)
```

```python
response["messages"][-1].content
```

```python
response = mem_agent.invoke({"messages": "what is my name"}, config=config)
```

```python
response
```

```python
response["messages"][-1].content
```

# Middleware

```python
from langchain.agents import create_agent
from langchain.agents.middleware import PIIMiddleware

agent = create_agent(
    model="groq:llama-3.3-70b-versatile",
    tools=[],
    middleware=[
            PIIMiddleware("email", strategy="redact", apply_to_input=True),
            PIIMiddleware("credit_card", strategy="mask", apply_to_input=True),
        ],
    
)
```

```python
agent.invoke({"messages":"my email is aaa@gmail.com"})
```
